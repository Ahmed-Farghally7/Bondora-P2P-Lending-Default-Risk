### Dataset Description 

Default risk has long been a significant risk factor to test borrowers’ behaviour in Peer-to-Peer (P2P) lending. In P2P lending, loans are typically uncollateralized and lenders seek higher returns as compensation for the financial risk they take. In addition, they need to make decisions under information asymmetry that works in favor of the borrowers. In order to make rational decisions, lenders want to minimize the risk of default of each lending decision and realize the return that compensates for the risk.

##### Let's remove features that could have potentially increased the problem of data leakage:
The data leakage can be introduced to the model if features that
would not be available in practice when the model is used for the prediction or the features65
that actually have the same information that the predictive model is trying to predict are used
to train the model. For this reason, all the features that are not available at the time when the investor is making
the investment decision (whether to invest in a certain loan or not), were excluded from the
final dataset. 
These variables included, for example, information related to payments during
the loan duration and the variables determining the overdue payments. The old versions of
Bondora’s credit ratings, the variables representing the estimated values for expected loss and
return of the loan and the variable indicating the probability of default were also removed from
the dataset to prevent the data leakage problem.

```python
df.drop(columns=['PlannedPrincipalTillDate', 'CurrentDebtDaysPrimary', 'PlannedPrincipalTillDate','CurrentDebtDaysPrimary',
                 'DebtOccuredOn', 'CurrentDebtDaysSecondary', 'ExpectedLoss', 'LossGivenDefault', 'ExpectedReturn', 
                 'ProbabilityOfDefault', 'PrincipalOverdueBySchedule','PlannedPrincipalPostDefault', 'PlannedInterestPostDefault',
                 'EAD1', 'EAD2', 'PrincipalRecovery','InterestRecovery', 'RecoveryStage', 'PrincipalWriteOffs', 'InterestAndPenaltyWriteOffs',
                 'GracePeriodStart', 'GracePeriodEnd', 'NextPaymentDate', 'NextPaymentNr', 'NrOfScheduledPayments',
                 'ReScheduledOn', 'PrincipalDebtServicingCost', 'InterestAndPenaltyDebtServicingCost', 'ActiveLateLastPaymentCategory',
                 'PlannedInterestTillDate', 'ActiveScheduleFirstPaymentReached', 'InterestAndPenaltyBalance', 'ActiveLateCategory',
                 'WorseLateCategory', 'InterestAndPenaltyPaymentsMade', 'PrincipalBalance', 'FirstPaymentDate', 'PrincipalPaymentsMade'], 
                  inplace = True )
```

###### - Let's perform feature scaling :

```python
def Scale(Column_name):
    # This is a function that takes a column name, performes data scaling, 
    # and then return the column after scaling its data 
    Column_name = (Column_name - min(Column_name)) / (max(Column_name) - min(Column_name))
    return Column_name
    
```

```python
# Now we can use Scale function to scale the numerical features
df.AppliedAmount = Scale(df.AppliedAmount)
df.Amount = Scale(df.Amount)
df.Interest = Scale(df.Interest)
df.MonthlyPayment = Scale(df.MonthlyPayment)
df.IncomeFromPrincipalEmployer = Scale(df.IncomeFromPrincipalEmployer)
df.IncomeFromPension = Scale(df.IncomeFromPension)
df.IncomeFromSocialWelfare = Scale(df.IncomeFromSocialWelfare)
df.IncomeFromLeavePay = Scale(df.IncomeFromLeavePay)
df.IncomeFromChildSupport  = Scale(df.IncomeFromChildSupport )
df.IncomeOther = Scale(df.IncomeOther)
df.IncomeTotal = Scale(df.IncomeTotal)
df.DebtToIncome = Scale(df.DebtToIncome)
df.FreeCash = Scale(df.FreeCash)
df.AmountOfPreviousLoansBeforeLoan = Scale(df.AmountOfPreviousLoansBeforeLoan)
df.PreviousRepaymentsBeforeLoan  = Scale(df.PreviousRepaymentsBeforeLoan)
df.PreviousEarlyRepaymentsBefoleLoan = Scale(df.PreviousEarlyRepaymentsBefoleLoan)
```
